```python
kbpedia = rdflib.Graph()
result = kbpedia.parse("kbpedia_reference_concepts.n3", format = "n3")
```

```python
LEN = len("http://kbpedia.org/kko/rc/")
kbpedia_subjects = [" ".join(re.findall("[A-z][a-z]*", subject[LEN:])).lower() for subject in kbpedia.subjects()]
```

```python
kbpedia_subjects[0]
```

```python
kbpedia_subjects[0:10]
```

```python
kbpedia_tokens = " ".join(kbpedia_subjects)
```

```python
kbpedia_vocabulary = nltk.word_tokenize(kbpedia_tokens)
```

```python
kbpedia_vocabulary = set(nltk.word_tokenize(kbpedia_tokens))
```

```python
len( vocabulary & kbpedia_vocabulary) / len(vocabulary) * 100
```
